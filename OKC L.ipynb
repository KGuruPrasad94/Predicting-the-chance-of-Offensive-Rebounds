## Calculating shot distance

```python
PID = Merge[['playerid_off_player_1', 'playerid_off_player_2',
                  'playerid_off_player_3', 'playerid_off_player_4',
                  'playerid_off_player_5']]
np_pid = PID.to_numpy()
np_pid
```

```python
np_pid.shape
```

```python
off_player_x_pos = Merge [['AtShot_loc_x_off_player_1', 'AtShot_loc_x_off_player_2',
                      'AtShot_loc_x_off_player_3', 'AtShot_loc_x_off_player_4',
                    'AtShot_loc_x_off_player_5']]
np_xpos = off_player_x_pos.to_numpy()
np_xpos
```

```python
np_xpos.shape
```

```python
off_player_y_pos = Merge [['AtShot_loc_y_off_player_1', 'AtShot_loc_y_off_player_2',
                      'AtShot_loc_y_off_player_3', 'AtShot_loc_y_off_player_4',
                    'AtShot_loc_y_off_player_5']]
np_ypos = off_player_y_pos.to_numpy()
np_ypos
```

```python
shooter_id_li = Merge['shooter_player_id'].values
shooter_id_li
```

```python
#finding shooter xpos and ypos
shooter_x_pos = []
shooter_y_pos = []

for i in range(0,len(shooter_id_li)):
    for j in range(0,len(PID.columns.values)):
        if(shooter_id_li[i]== np_pid[i][j]):
            shooter_x_pos.append(np_xpos[i][j])
            shooter_y_pos.append(np_ypos[i][j])
            
```

```python
Shooter_pos = pd.DataFrame()
Shooter_pos['shooter_x_pos'] = shooter_x_pos
Shooter_pos['shooter_y_pos'] = shooter_y_pos
#Shooter_pos_dr = Shooter_pos_dr[Shooter_pos_dr['shooter_x_pos_dr']<0]
#Shooter_pos_dr
Shooter_pos.head()
```

```python
Shooter_pos.shape
```

```python
Shooter_pos.isnull().sum()
```

```python
#calculating shot distance from rim
Shooter_pos['Shot_distance'] = ((Shooter_pos['shooter_x_pos'] - (-41.75)).pow(2) + (Shooter_pos['shooter_y_pos'] - 0).pow(2)).pow(1/2)
Shooter_pos['Shot_distance'] 
```

```python
#calculating shot distance from rim
Merge['Shot_distance'] = ((Shooter_pos['shooter_x_pos'] - (-41.75)).pow(2) + (Shooter_pos['shooter_y_pos'] - 0).pow(2)).pow(1/2)
Merge['Shot_distance'] = Merge['Shot_distance'].fillna(Merge['Shot_distance'].mean())
```

## Resampling data

```python
from sklearn.utils import resample
# Separate majority and minority classes
df_majority1 = Merge[Merge['f.oreb']==2]
df_majority2 = Merge[Merge['f.oreb']==1]
df_minority = Merge[Merge['f.oreb']==0]
 
# Downsample majority1 class
df_majority1_downsampled = resample(df_majority1, 
                                 replace=False,    # sample without replacement
                                 n_samples=46887,     # to match minority class
                                 random_state=123) # reproducible results

df_majority2_downsampled = resample(df_majority2, 
                                 replace=False,    # sample without replacement
                                 n_samples=46887,     # to match minority class
                                 random_state=123) # reproducible results 

# Combine minority class with downsampled majority class
downsampled = pd.concat([df_majority1_downsampled,df_majority2_downsampled, df_minority])
 
# Display new class counts
downsampled['f.oreb'].value_counts()
```
